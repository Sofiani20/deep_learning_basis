iris contient 3 espèces de fleur en fonction de la largeur et la longueur de leurs sépales et pétales.

* 0 pour **setosa**
* 1 pour **versicolor**
* 2 pour **virginica**

sepal lentg || sepal widht || petal length || petal widht

#### Nous voulons utiliser les variables petal length et width pour prédir si la fleur est de l'espèce setosa ou pas en utilisant un perceptron sepal

```python
X=iris.data[:,(2,3)]
```

####  y=1 si setosa y=0 sinon

```python
y=iris.target
```

```python
y
```

```python
y[y==1]=3
```

```python
y[y==0]=1
```

```python
y
```

```python
y[y!=1]=0
```

```python
y
```

```python
X
```

```python
from sklearn.linear_model import Perceptron
```

```python
help(Perceptron)
```

```python
per_clf=Perceptron(random_state=42)
per_clf.fit(X,y)
```

```python
y_pred=per_clf.predict(X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y,y_pred)
```

```python
print(per_clf.coef_)
print(per_clf.intercept_)
```
